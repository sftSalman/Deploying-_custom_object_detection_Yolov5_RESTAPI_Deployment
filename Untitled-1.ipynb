```python
import os
import shutil
import glob
import random


def preprocessing(image_folder, annotation_folder):
    # Create the merged folder
    merged_folder = os.path.join(os.getcwd(), "merged")
    os.makedirs(merged_folder, exist_ok=True)

    # Check the extension of images and convert them into JPG format
    for img_path in glob.glob(os.path.join(image_folder, "*")):
        if img_path.endswith((".jpg", ".jpeg", ".png")):
            img_name = os.path.basename(img_path)
            img_jpg_path = os.path.join(merged_folder, img_name.split(".")[0] + ".jpg")
            shutil.copy(img_path, img_jpg_path)

    # Merge the annotation folder and image folder into a single folder called merged
    for ann_path in glob.glob(os.path.join(annotation_folder, "*")):
        ann_name = os.path.basename(ann_path)
        img_name = ann_name.split(".")[0] + ".jpg"
        img_path = os.path.join(merged_folder, img_name)
        if os.path.exists(img_path):
            ann_merged_path = os.path.join(merged_folder, ann_name)
            shutil.copy(ann_path, ann_merged_path)

    # Split the images and annotations from merged folder into obj and test for YOLOv4
    with open("obj_train.txt", "w") as obj_train_file, open("obj_test.txt", "w") as obj_test_file:
        for img_path in glob.glob(os.path.join(merged_folder, "*.jpg")):
            ann_path = os.path.join(merged_folder, os.path.basename(img_path).split(".")[0] + ".txt")
            if os.path.exists(ann_path):
                # Move the annotation file and image file to obj or test directory
                if random.random() < 0.8:
                    shutil.move(ann_path, "obj")
                    shutil.move(img_path, "obj")
                    obj_train_file.write(os.path.abspath(img_path) + "\n")
                else:
                    shutil.move(ann_path, "test")
                    shutil.move(img_path, "test")
                    obj_test_file.write(os.path.abspath(img_path) + "\n")

    # Remove the merged folder
    

    print("Preprocessing completed successfully!")
```

```python
import os
import shutil
import glob
import random

def preprocessing(image_folder, annotation_folder):
    # Create the new_merged folder
    new_merged_folder = os.path.join(os.getcwd(), "new_merged")
    os.makedirs(new_merged_folder, exist_ok=True)

    # Check the extension of images and convert them into JPG format
    for img_path in glob.glob(os.path.join(image_folder, "*")):
        if img_path.endswith((".jpg", ".jpeg", ".png")):
            img_name = os.path.basename(img_path)
            img_jpg_path = os.path.join(new_merged_folder, img_name.split(".")[0] + ".jpg")
            shutil.copy(img_path, img_jpg_path)

    # Merge the annotation folder and image folder into a single folder called new_merged
    for ann_path in glob.glob(os.path.join(annotation_folder, "*")):
        ann_name = os.path.basename(ann_path)
        img_name = ann_name.split(".")[0] + ".jpg"
        img_path = os.path.join(new_merged_folder, img_name)
        if os.path.exists(img_path):
            ann_merged_path = os.path.join(new_merged_folder, ann_name)
            shutil.copy(ann_path, ann_merged_path)

    # Create obj and test directories
    os.makedirs("obj", exist_ok=True)
    os.makedirs("test", exist_ok=True)

    # Split the images and annotations from new_merged folder into obj and test for YOLOv4
    with open("obj_train.txt", "w") as obj_train_file, open("obj_test.txt", "w") as obj_test_file:
        for img_path in glob.glob(os.path.join(new_merged_folder, "*.jpg")):
            ann_path = os.path.join(new_merged_folder, os.path.basename(img_path).split(".")[0] + ".txt")
            if os.path.exists(ann_path):
                # Move the annotation file and image file to obj or test directory
                if random.random() < 0.8:
                    shutil.move(ann_path, "obj")
                    shutil.move(img_path, "obj")
                    obj_train_file.write(os.path.abspath(img_path) + "\n")
                else:
                    shutil.move(ann_path, "test")
                    shutil.move(img_path, "test")
                    obj_test_file.write(os.path.abspath(img_path) + "\n")
                    
    # Create classes.txt file in obj and test
    with open(os.path.join("obj", "classes.txt"), "w") as classes_file:
        classes_file.write("cat\n")
        classes_file.write("dog\n")
    with open(os.path.join("test", "classes.txt"), "w") as classes_file:
        classes_file.write("cat\n")
        classes_file.write("dog\n")

    # Check if annotations are done correctly
    obj_ann = glob.glob(os.path.join("obj", "*.txt"))
    test_ann = glob.glob(os.path.join("test", "*.txt"))
    obj_no_ann = len(os.listdir("obj")) - len(obj_ann)
    test_no_ann = len(os.listdir("test")) - len(test_ann)
    if obj_no_ann > 0:
        print(f"Warning: {obj_no_ann} images in obj have no annotations.")
    if test_no_ann > 0:
        print(f"Warning: {test_no_ann} images in test have no annotations.")
    
    # Print the number of obj and test data
    print(f"Number of images in obj: {len(os.listdir('obj'))}")
    print(f"Number of images in test: {len(os.listdir('test'))}")
    print(f"Preprocessing completed successfully!")
```
